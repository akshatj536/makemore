```python
g=torch.Generator().manual_seed(14526)

block_size = 3
X, Y = [],[]
for w in words[:]:
    #print(w)
    context = [0]* block_size
    for ch in w +'.':
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        #print(''.join(itos[i] for i in context), '--->', itos[ix])
        context = context[1:] + [ix] 
X= torch.tensor(X)
Y= torch.tensor(Y)
```

```python
g=torch.Generator().manual_seed(2147483647)
W1 = torch.randn((6,100), requires_grad= True, generator=g)
b1 = torch.randn(100, requires_grad=True, generator=g)
W2 = torch.randn((100,27), requires_grad= True, generator=g)
b2 = torch.randn((27), requires_grad= True, generator=g)
C=torch.randn((27,2), requires_grad= True, generator=g)
parameters=[C,W1,b1,W2,b2]
```

```python
emb= C[X]
emb.shape
```

```python
for _ in range(100):
    # forward
    h = emb.view(228146, 6) @ W1 + b1
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y)

    # backward
    loss.backward()

    
    
    for p in parameters:
        p -= 0.1 * p.grad   # ✅ this works ONLY inside no_grad
    
    print(loss.item())
    # clear grads for next step
    for p in parameters:
        p.grad = None   # ✅ modern way (same as zero_())

    print(loss.item())
```
